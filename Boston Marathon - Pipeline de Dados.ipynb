### Principais problematicas encontradas e soluções:

######1. Atributos divergentes entre as tabelas. 
Solução: criar dataframes a partir desses apenas com os dados necessários para analise dos objetivos

######2. Formato do tempo divergente.
Comparação de tempo ficará inviavel uma vez que os formatos não batem
Solução: Transformar os formatos para segundos

######3. Juntar os dados por ano
Embora não seja necessário, facilitará o entendimentos e as analises dos objetivos
Solução: Adicionar um atributo que separe por ano (2023/2024)

######4. Formato do genero divergente
Enquanto em 2024 está com M/F (Male/Masculino, Female/Feminimo) em 2023 está com M/W (Man/Homem, Woman/Mulher). 
Solução: Onde tiver W, substituir por F para manter o padrão.

######5. Outliers de genero no df de 2024.
Existe um Genero classificado como "X"
Solução: Uma vez que não existe uma classificação dentro de um genero indefinido/não binario e como são poucos os registros (menos de 1% do total), optou-se por excluir esses registros para não prejudicar a contagem final

######6. Idade não está separada por Grupos de Idade nos dados de 2024
Solução: Criar um atributo que distribua as idades em grupos.

```
#imports necessarios

from pyspark.sql.functions import when,split, concat_ws, trim,col, regexp_extract, format_string, floor, coalesce, lit
```

# Soluções:

### 1. Ajustando os dataframes para tirar atributos irrelevantes

```
df_2024.columns
```

```
df_2024 = df_2024.select("FullName","BibNumber", "AgeOnRaceDay", "gender","CountryofResAbbrev","OfficialTime","RankOverall","RankOverGender")
df_2024 = df_2024.toDF("nome", "numero_peito", "idade", "genero", "país", "tempo_final","pos_final", "pos_genero")
df_2024.show(10)
```

```
df_2023.columns
```

```
df_2023 = df_2023.select("name","bib_number", "age_group", "gender","finish_net","place_overall","place_gender")
df_2023 = df_2023.toDF("nome", "numero_peito", "faixa_etaria", "genero", "tempo_final","pos_final", "pos_genero")

df_2023 = df_2023.withColumn("nome", concat_ws(" ",
    trim(split(col("nome"), ",").getItem(1)),
    trim(split(col("nome"), ",").getItem(0))
))

df_2023.show(10)
```

### 2. Padronizando os atributos de tempo final 

```

df_2023 = df_2023.withColumn("horas", coalesce(regexp_extract("tempo_final", r"(\d+)H", 1).cast("int"), lit(0))) \
                 .withColumn("minutos", coalesce(regexp_extract("tempo_final", r"(\d+)M", 1).cast("int"), lit(0))) \
                 .withColumn("segundos", coalesce(regexp_extract("tempo_final", r"(\d+)S", 1).cast("int"), lit(0)))

# Tempo em segundos
df_2023 = df_2023.withColumn("tempo_final_segundos", 
col("horas") * 3600 + col("minutos") * 60 + col("segundos"))


df_2023 = df_2023.withColumn(
    "tempo_final",
    format_string(
        "%d:%02d:%02d",
        floor(col("tempo_final_segundos") / 3600),
        floor((col("tempo_final_segundos") % 3600) / 60),
        col("tempo_final_segundos") % 60
    )
)

df_2023 = df_2023.drop("horas", "minutos", "segundos","tempo_final_segundos")
df_2023.show(10)
```

###3. Adicionar o ano aos dataframes


```
from pyspark.sql.functions import lit
df_2023 = df_2023.withColumn("ano", lit(2023))
df_2024 = df_2024.withColumn("ano", lit(2024))
```

###4. Ajustar o genero (de W pra F)

```
%sql

select * from df_2023
where genero = "W"
```

```
#Views pra ver dados com SQL

df_2023.createOrReplaceTempView("df_2023")
df_2024.createOrReplaceTempView("df_2024")
```

```
df_2023 = df_2023.withColumn(
    "genero",
    when(col("genero") == "W", "F").otherwise(col("genero"))
)
```

```
df_2023.createOrReplaceTempView("df_2023")
```

```
%sql

select * from df_2023
where genero = "F"
```

###5. Ajustar genero indeterminado (X)

```
#atualiza a view pra poder ver os dados atualizados
df_2024.createOrReplaceTempView("df_2024")
```

```
%sql

SELECT *
from df_2024
where genero = "X"
```

```
df_2024 = df_2024.filter(col("genero") != "X")
df_2024.createOrReplaceTempView("df_2024")
```

###6. Ajustar grupos de idade

```
df_2024 = df_2024.withColumn(
    "faixa_etaria",
    when(col("idade") < 18, "Under 18")
    .when((col("idade") >= 18) & (col("idade") <= 39), "18-39")
    .when((col("idade") >= 40) & (col("idade") <= 44), "40-44")
    .when((col("idade") >= 45) & (col("idade") <= 49), "45-49")
    .when((col("idade") >= 50) & (col("idade") <= 54), "50-54")
    .when((col("idade") >= 55) & (col("idade") <= 59), "55-59")
    .when((col("idade") >= 60) & (col("idade") <= 64), "60-64")
    .when((col("idade") >= 65) & (col("idade") <= 69), "65-69")
    .when((col("idade") >= 70) & (col("idade") <= 74), "70-74")
    .when((col("idade") >= 75) & (col("idade") <= 79), "75-79")
    .when(col("idade") >= 80, "80+")
    .otherwise("INVALIDO")
)
```

```
df_2024 = df_2024.drop ("idade")
```

```
df_2024.show(10)
```

```
%sql

select * from df_2024 where idade = "INVALIDO"
```

# Tabelas Fato e Dimensão

```
#Separar as colunas em fato e dim

from pyspark.sql.functions import monotonically_increasing_id
```

```
df_maratona = df_2024.unionByName(df_2023, allowMissingColumns=True)
df_maratona.show()
```

```
#dim participante

dim_participante = df_maratona.select("nome", "numero_peito", "tempo_final").distinct()

# id participante
dim_participante = dim_participante.withColumn("id_participante", monotonically_increasing_id())
dim_participante.show(5)
```

```
#dim categoria

dim_categoria = df_maratona.select("genero", "faixa_etaria").distinct()
dim_categoria = dim_categoria.withColumn("id_categoria", monotonically_increasing_id())

dim_categoria.show(5)
```

```
#fato maratona

df_fato = df_maratona.join(dim_participante, ["nome", "numero_peito", "tempo_final"], "left")
df_fato = df_fato.join(dim_categoria, ["genero", "faixa_etaria"], "left")

fato_maratona = df_fato.select(
    monotonically_increasing_id().alias("id_fato"),
    "ano",
    "id_participante",
    "id_categoria",
    "pos_final",
    "pos_genero"
)

fato_maratona.show(5)
```

### Objetivo 1:

Comparar o numero de inscritos entre os anos de 2023 e 2024 na Maratona de Boston para entender se realmente houve um aumento no volume de inscrições

```
%sql
select ano, count(*) as total_participantes from fato_maratona
group by ano
```

###Nota sobre objetivo 1:
Honestamente, esse resultado foi surpreendente. Esperava um aumento nas inscrições. Porém isso pode ser explicado pelo fato de que a maratona de Boston exige um indice (tempo maximo que um atleta tem que ter, comprovadamente,concluido uma maratona, documentado em uma prova oficial) para entrada sem sorteio.

Em 2023, esse indice, para faixa etaria de 18 a 39 anos, era de 3 horas. Em 2024, foi de 2h52minutos, exigindo que para se qualificar, os atletas provassem que correm 1km em aproximadamente 4 minutos. A diferença parece pequena, mas em longos percursos, é uma alta velocidade.

Embora existam outros meios, os mesmos são mais custosos financeiramente. Portanto, pode-se esperar que esse seja um fator que fez diferença nas inscrições.

### Objetivo 2:
Entender se, comparativamente, houve diferença na porcentagem dos inscritos por genero em cada ano

```
%sql
  SELECT f.ano,d.genero, COUNT(*) AS total
  FROM fato_maratona f
  JOIN dim_categoria d ON f.id_categoria = d.id_categoria
  GROUP BY f.ano, d.genero
  order by d.genero, f.ano
```

###Nota sobre objetivo 2:
Como esperado pela diminuição de inscrições, em ambos os generos o numero de participantes caiu.
Comparado com 2023, 2024 teve uma diminuição de 4.6% de participantes mulheres e 4% de participantes homens. Porém, comparativamente, houve um ligeiro aumento na porcentagem de mulheres inscritas em 2024, indo de 42.7% no ano de 2023 para 43% no ano de 2024. Pode-se concluir então que mesmo com uma tendencia a diminuir as incrições, houve uma igualdade ligeiramente maior na questão de genero no ano de 2024.

### Objetivo 3
•	Entender se, comparativamente, houve diferença nos inscritos por faixa etaria

```
%sql
SELECT f.ano,d.faixa_etaria,COUNT(*) AS total
  FROM fato_maratona f
  JOIN dim_categoria d ON f.id_categoria = d.id_categoria
  GROUP BY f.ano, d.faixa_etaria
  ORDER BY d.faixa_etaria, f.ano
```

###Nota sobre objetivo 3:
De forma parecida com o objetivo 2, há alguns pontos interessantes a serem observados.

Até os 54 anos, todas as faixas tiveram quedas consideraveis, porém, a partir dos 55, acontece um fato interessante e o movimento é o contrario: todas as inscrições subiram.  A faixa etaria de 80+, em especial, teve um aumento de 100% sua participação, indo de 5 para 11 pessoas

Isso ainda pode ser explicado pelo índice exigido. Como a cada faixa etaria esse tempo aumenta, vê-se que nas faixas mais jovens é mais inalcansavel do que para os mais velhos.

### Objetivo 5
Analisar a media de tempos em cada grupo por faixa etaria e genero para entender se houve melhora ou piora entre anos

```
%sql
SELECT f.ano,cat.genero, cat.faixa_etaria, ROUND(AVG(part.tempo_segundos) / 60, 2) AS media_tempo_minutos
FROM fato_maratona f
JOIN dim_participante part ON f.id_participante = part.id_participante
JOIN dim_categoria cat ON f.id_categoria = cat.id_categoria
GROUP BY f.ano, cat.genero, cat.faixa_etaria
ORDER BY cat.genero, cat.faixa_etaria, f.ano
```

###Nota sobre objetivo 5:

Novamente, entre os aspectos, existem observações importantes de se fazer.
No geral, a porcentagem da media de tempo caiu pouco (entre 3 e 5% a mais que 2023) para o genero feminino nas categorias mais jovens, até seus 55 anos. Porém nas categorias mais idosas, a porcentagem aumentou muito, chegando aos 15% nos 80+. O contrario se observou nas categorias masculinas, onde os mais jovens tiveram uma variação mais alta (entre 5 e 7% a mais que em 2023), porém observando as idades mais avançadas, o crescimento é minimo (1,8% apenas na faixa entre 75 e 79 anos em comparação com 2023) e até chega, na categoria masculina 80+, a ser 0.31% mais rapida que a media de 2023
